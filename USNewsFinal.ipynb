```python
import requests
from splinter import Browser
from bs4 import BeautifulSoup
import pymongo

import time

executable_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **executable_path)
url = 'https://realestate.usnews.com/places/rankings/best-places-to-retire'
browser.visit(url)

headers = {"User-agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.80 Safari/537.36"}
   
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text,'html')

for i in range(9):
    try:
        browser.find_link_by_text('Load More').first.click()
        time.sleep(5)
    except ValueError:
        print("Skipping...")
        
        
time.sleep(10)
html = browser.html
soup = BeautifulSoup(html, 'lxml')
results = soup.find_all('div', class_='column small-12 medium-8')

top100=[]
init_list=[]
i=0
for result in results:
    if i<=99:
        rank=i
        citystate=result.find("a").get_text()
        city=citystate.split(",")[0]
        state=citystate.split(",")[1]
        x=0
        for res in results[i]:
            if x==7:
                img_url=res.img["src"]
            x=x+1
        dupctr=0
        for lst in top100:
            if lst["city"]==city:
                dupctr=1
               
        if dupctr==0:
            top100.append({"rank":rank,"city":city,"state":state,"url":img_url})
    i=i+1
    
print(top100)
```

```python
import json
with open ("./Datasets/USNews.json","w") as outfile:
    json.dump(top100,outfile)
```
